## Importing libraries needed


```python
import gym
import keras as K
import numpy as np
import tensorflow as tf
from PIL import Image
from rl.agents.dqn import DQNAgent
from rl.memory import SequentialMemory
from rl.policy import EpsGreedyQPolicy
from rl.policy import LinearAnnealedPolicy
from rl.processors import Processor
```

## Training script

```python
def make_model(num_actions: int=4) -> K.models.Model:
    """
    creates blueprint for both, the model needed for the agent's actions
    and the other model used to predict the future rewards

    it turns out in an approximator of the Q-Learning algorithm
    with the objective of converging on it
    
    this network is defined in the paper:
    https://arxiv.org/pdf/1312.5602.pdf

    - num_actions: num of actions which the agent would take
    Returns: the model needed
    """
    # the 4 is because we take 4 frames per decision
    # to determine not just position but also direction and velocity
    x = K.layers.Input(shape=(4, 84, 84))
    y = K.layers.Permute(dims=(2, 3, 1))(x)

    y = K.layers.Conv2D(filters=32, kernel_size=(8, 8), strides=(4, 4),
                        activation='relu', data_format='channels_last')(y)    
    y = K.layers.Conv2D(filters=64, kernel_size=(4, 4), strides=(2, 2),
                        activation='relu', data_format='channels_last')(y)
    y = K.layers.Conv2D(filters=64, kernel_size=(3, 3), strides=(1, 1),
                        activation='relu', data_format='channels_last')(y)
    
    y = K.layers.Flatten()(y)

    y = K.layers.Dense(units=512, activation='relu')(y)
    y = K.layers.Dense(units=4, activation='linear')(y)
    
    return K.Model(inputs=x, outputs=y)
```

```python
class ProcessAtari(Processor):
    """
    preprocess data
    """

    def process_observation(self, batch: np.ndarray):
        """
        preprocess in format described in
        Playing Atari with Deep Reinforcement Learning paper
        """
        if not isinstance(batch, np.ndarray):
            raise ValueError('batch must be a np.ndarray')
        if len(batch.shape) != 3:
            raise ValueError('batch must contain 3 dimensions')
        batch = Image.fromarray(batch)
        batch = batch.resize((84, 84), Image.ANTIALIAS).convert('L')
        return np.array(batch).astype('uint8')

    def process_state_batch(self, batch):
        """
        normalizes images
        """
        return batch.astype('float32') / 255.
    
    def process_reward(self, reward):
        """
        process reward
        """
        return np.clip(reward, -1., 1.)
```

```python
def train(gamma: float=.99, epsilon: float=1.,
          max_steps_per_episode: int=50000, nb_actions: int=4) -> tuple:
    """
    trains an agent which plays atari breakout
    - gamma: discount factor for past rewards
    - epsilon: epsilon greedy's threshold
    - epsilon_min: min value epsilon must take
    - epsilon_max: max value epsilon must take
    - epsilon_interval: max - min (distance) between those intervals
    - max_steps_per_episode: maximum quantity of steps per episode
    - nb_actions: number of actions the agent would take
    Returns: a tuple with the training history
           and the path in which the training was saved
    """
    env = gym.make('BreakoutDeterministic-v4')
    env.reset()
    memory = SequentialMemory(limit=1000000, window_length=nb_actions)
    policy = LinearAnnealedPolicy(EpsGreedyQPolicy(),
                                attr='eps',
                                value_max=epsilon,
                                value_min=.1,
                                value_test=.05,
                                nb_steps=1000000)
    model = make_model()
    agent = DQNAgent(model=model, memory=memory,
                     policy=policy,
                     nb_actions=nb_actions, nb_steps_warmup=max_steps_per_episode,
                     target_model_update=10000,
                     gamma=gamma,
                     delta_clip=1.,
                     processor=ProcessAtari(),
                     train_interval=4)
    agent.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=.00025),
                  metrics=['mae'])
    agent.fit(env=env, nb_steps=max_steps_per_episode, verbose=2)
    agent.save_weights(filepath='policy.h5',
                       overwrite=True)
```

```python
train()
```

## Now is time to play

```python
def play(path: str='policy.h5', nb_actions: int=4):
    """
    function which makes play breakout to the trained agent
    - path: file location
    - nb_actions: number of actions agent can take
    """
    model = make_model()
    memory = SequentialMemory(limit=1000000, window_length=nb_actions)
    agent = DQNAgent(model=model, memory=memory,
                     nb_actions=nb_actions,
                     processor=ProcessAtari())
    agent.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=.00025),
                  metrics=['mae'])
    agent.load_weights(path)
    
    env = gym.make('BreakoutDeterministic-v4')
    env.reset()
    agent.test(env, nb_episodes=50, visualize=True)
```

```python
play()
```
